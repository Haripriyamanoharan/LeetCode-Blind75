```python
https://leetcode.com/problems/merge-two-sorted-lists/description/?envType=study-plan&id=level-1


# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        temp1=list1
        temp2=list2
        temp4=ListNode()
        temp3=ListNode()
        temp4=temp3
        if(temp1 is None and temp2 is None):
            return None
        elif(temp1 is None):
            return temp2
        elif(temp2 is None):
            return temp1
        while(temp1 and temp2 ):
            if(temp1.val==temp2.val):
                temp3.next=ListNode(temp1.val)
                temp3=temp3.next
                temp3.next=ListNode(temp2.val)
                temp3=temp3.next
                temp1=temp1.next
                temp2=temp2.next
            elif(temp1.val<temp2.val):
                temp3.next=ListNode(temp1.val)
                temp3=temp3.next
                if(temp1.next is None or temp2.val < temp1.next.val):
                    temp3.next=ListNode(temp2.val)
                    temp3=temp3.next
                    temp2=temp2.next
                temp1=temp1.next
            elif(temp1.val>temp2.val):
                temp3.next=ListNode(temp2.val)
                temp3=temp3.next
                if(temp2.next== None or temp1.val < temp2.next.val):
                    temp3.next=ListNode(temp1.val)
                    temp3=temp3.next
                    temp1=temp1.next
                temp2=temp2.next
        temp=None
        if(temp1 is None and temp2 is not None):
            temp=temp2
        elif(temp2 is None and temp1 is not None):
            temp=temp1
        while(temp):
            temp3.next=temp
            temp=temp.next
            temp3=temp3.next
        return temp4.next
```
